```python
import numpy as np

L=10;T=5;p=0.4
road=np.random.choice([0,1],size=L,p=[1-p,p])
total_cars=np.sum(road)
print("Initial road:",''.join(map(str,road)))
print("Total cars:",total_cars)
for step in range(T):
    new_road=np.zeros_like(road);moves=0
    for i in range(L):
        if road[i]==1:
            n=(i+1)%L
            if road[n]==0:
                new_road[n]=1;moves+=1
            else:new_road[i]=1
    print(f"Step {step+1:02d}: {''.join(map(str,new_road))} | Cars moved: {moves}")
    road=new_road.copy()
print("Final road:",''.join(map(str,road)))
print("Total cars:",total_cars)
```
